```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # For heatmaps
```

```python
# Step 2.1: Create matrices (YOUR sales data example)
# Rows = Days, Columns = [Revenue, Customers, Units]
```

```python
day1_sales=np.array([[1000,50,200], [1200, 60, 250]])  # 2 days x 3 metrics
day2_sales = np.array([[1100, 55, 220], [1300, 65, 270], [900, 45, 180]])   # 3 days x 3 metrics

print("Day1 (2x3):\n", day1_sales)
print("Day2 (3x3):\n", day2_sales)
```

```python
#verify if multiplication between A and B is possible A(2x{3}) , B({3}x3) , it gives C(2,3)
print("Compatible? Day1 cols (3) == Day2 rows (3):", day1_sales.shape[1]   == day2_sales.shape[0]) 
# day1_sales.shape[1]  : rows shape
# day2_sales.shape[0]  : cols shape
```

```python
# Step 2.2: Matrix Multiplication (Daily totals aggregation)
total_sales=np.dot(day1_sales,day2_sales)
total_sales
```

```python
# Step 2.3: NumPy @ operator (modern way)
total_sales_V2= day1_sales @ day2_sales
total_sales_V2
```

```python
# Step 2.4: Visualize HEATMAP (Business dashboard style)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); sns.heatmap(day1_sales, annot=True, cmap='Blues'); 
plt.title('Day1 Sales')
plt.subplot(1,3,2); sns.heatmap(day2_sales, annot=True, cmap='Greens'); 
plt.title('Day2 Sales')
plt.subplot(1,3,3); sns.heatmap(total_sales, annot=True, cmap='Reds'); 
plt.title('Aggregated Totals')
plt.tight_layout(); plt.show()
```
