```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2.1: Create YOUR fitness vector (sleep hrs, steps, cals)
your_day = np.array([7.5, 12000, 2500])  # Customize!
friend_day = np.array([8.0, 11000, 2200])

print("Your vector:", your_day)
print("Friend's vector:", friend_day)
```

```python
# Step 2.2: Magnitude (length/energy)
your_mag = np.linalg.norm(your_day)
print(f"Your day magnitude: {your_mag:.2f}")
```

```python
# Step 2.3: Dot product (similarity score)
similarity = np.dot(your_day, friend_day)
print(f"Similarity: {similarity:.2f}")
```

```python
# Step 2.4: Visualize (arrow plot)
plt.figure(figsize=(8,6))
plt.quiver(0, 0, your_day[0], your_day[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Your Day')
plt.quiver(0, 0, friend_day[0], friend_day[1], angles='xy', scale_units='xy', scale=1, color='red', label="Friend's Day")
plt.xlim(0, 9); plt.ylim(0, 13000)
plt.xlabel('Sleep (hrs)'); plt.ylabel('Steps')
plt.title('Vector Similarity: Fitness Days'); plt.legend(); plt.grid(); plt.show()
```

```python
# shows the angle between 2 vectors
theta = np.arccos(similarity / (your_mag * np.linalg.norm(friend_day)))
theta
```
