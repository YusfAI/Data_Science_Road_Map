```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats  # Hypothesis testing!
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
# Step 2.1: Generate data (your fitness: steps before/after new routine)
old_routine = np.random.normal(10000, 2000, 100)  # Mean 10k, std 2k
new_routine = np.random.normal(12000, 1800, 100)  # Better?
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.hist(old_routine, bins=20, alpha=0.7, label='Old'); plt.title('Distributions')
plt.subplot(1,3,2); plt.hist(new_routine, bins=20, alpha=0.7, label='New', color='green'); plt.title('New Routine')
plt.subplot(1,3,3); sns.boxplot(data=[old_routine, new_routine]); plt.title('Boxplot Compare')
plt.tight_layout(); plt.show()
```

```python
print(f"Old: Mean={old_routine.mean():.0f}, Std={old_routine.std():.0f}")
print(f"New: Mean={new_routine.mean():.0f}, Std={new_routine.std():.0f}")
```

```python
# Step 2.2: T-Test (Is new > old? p<0.05 = significant!)
t_stat, p_value = stats.ttest_ind(new_routine, old_routine, equal_var=False)
print(f"\nT-test: t={t_stat:.2f}, p={p_value:.4f}")
print("Significant?" , "YES! 🔥" if p_value < 0.05 else "No (noise)")
```
